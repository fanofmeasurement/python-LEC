```python
class poor(object):
    def __init__(self,education,house,water,mems):
        self.education=education
        self.house=house
        self.water=water
        self.mems=mems
    def really_poor(self):
        self.sum = 0
        if self.education == 0:
            self.sum += 1
        if self.house == 0:
            self.sum += 1
        if self.water == 0:
            self.sum += 1
        if self.mems == 0:
            self.sum += 1
        if self.sum > 2:
            print('æ­¤æˆ·äººå®¶ä¸ºè´«å›°æˆ·')
        else:
            print("æ­¤æˆ·äººå®¶ä¸å±äºè´«å›°æˆ·")
    def help_poor(self):
        if self.sum>2:
            if self.education==0:
                print('ç»™äºˆä¹‰åŠ¡æ•™è‚²å®‰å…¨ä¿éšœ,å¸®æ‰¶é€‰æ‹©1ï¼Œä¸å¸®æ‰¶ä¸º0')
                self.education=int(input())
                if self.education==1:
                    self.sum-=1
            if self.house==0:
                print('ç»™äºˆä½æˆ¿å®‰å…¨ä¿éšœ,å¸®æ‰¶é€‰æ‹©1ï¼Œä¸å¸®æ‰¶ä¸º0')
                self.house=int(input())
                if self.house==1:
                    self.sum-=1
            if self.water==0:
                print('ç»™äºˆé¥®æ°´å®‰å…¨ä¿éšœ,å¸®æ‰¶é€‰æ‹©1ï¼Œä¸å¸®æ‰¶ä¸º0')
                self.water=int(input())
                if self.water==1:
                    self.sum-=1
            if self.mems==0:
                print('ç»™äºˆè‚‰è›‹å¥¶è±†å®‰å…¨ä¿éšœ,å¸®æ‰¶é€‰æ‹©1ï¼Œä¸å¸®æ‰¶ä¸º0')
                self.mems=int(input())
                if self.mems==1:
                    self.sum-=1
            if self.sum>2:
                print("æ‰¶è´«ä¸æˆåŠŸ")
            else:
                print("æ‰¶è´«æˆåŠŸ")

family1=poor(0,0,0,1)
print('ç¬¬ä¸€ä¸ªå®¶åº­:')
family1.really_poor()
family1.help_poor()
family2=poor(1,1,0,1)
print('\nç¬¬äºŒä¸ªå®¶åº­:')
family2.really_poor()
family2.help_poor()
family3=poor(0,1,0,0)
print('\nç¬¬ä¸‰ä¸ªå®¶åº­:')
family3.really_poor()
family3.help_poor()
```

```python
class student(object):
    def __init__(self,name,score,sleep_time):
        self.name=name
        self.score=score
        self.sleep_time=sleep_time
    def predict(self,score,sleep_time):
        print('è¯·è¾“å…¥åˆ†æ•°æ ‡å‡†,å®è¡Œç™¾åˆ†åˆ¶')
        standard_score=int(input())
        print('è¯·è¾“å…¥ç¡çœ æ—¶é—´æ ‡å‡†,å…¶å•ä½ä¸ºå°æ—¶')
        standard_sleep_time=int(input())
        if (self.score > standard_score) & (self.sleep_time<standard_sleep_time):
            print ('BMIå¯èƒ½åå¤§')
        else:
            print('ä½“é‡æ­£å¸¸')

man1=student('å°æ—å­-æœ€å¸…çš„Makiå‰å¨ƒå¨ƒğŸ¶',100,8)
print("è¥ä¸šéƒ¨ç¬¬ä¸€åå¥½å˜›:")
man1.predict(score=100,sleep_time=8)

man2=student('ç”œåœ­-äººç”Ÿèµ¢å®¶æ˜¥ç”°ç”°',60,3)
print("\nemmm,ä¸šç»©å°±ä¸è¯´äº†,è¿˜ç†¬å¤œçœ‹çƒï¼Ÿï¼Ÿï¼Ÿ")
man2.predict(score=60,sleep_time=3)

man3=student('æ”¿å®—ä¸»ä»»-åªèƒ½æ˜¯çˆ±æƒ…è‹¦æƒ…å‰§ä¸­Makiçš„å‰ç”·å‹äº†',85,9)
print("\næ´ç™–ä¸»ä»»ğŸŒšğŸŒšğŸŒš:")
man3.predict(score=85,sleep_time=9)

man4=student('èˆé¦™ã•ã‚“å…¶å®æ˜¯ä¸ªç‰§æ˜¥å¥³å­©',70,6)
print("\nå…«å¦ä¼ é€æœºğŸŒšğŸŒšğŸŒš:")
man4.predict(score=70,sleep_time=6)
```
