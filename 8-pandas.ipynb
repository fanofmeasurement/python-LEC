## pandas

Pandas提供DataFrame数据结构，与R中的data.frame类似。内含常用的分组统计，缺失值处理等操作，包括DataFrame的读写和各种类型的转换

```python
import pandas as pd
```

```python
from pandas import Series, DataFrame
```

## series 

```python
obj = pd.Series([4, 7, -5, 3])
obj
```

```python
obj.values
```

```python
obj.index  # like range(4)
```

左边表示index，右边表示对应的value。可以通过value和index属性查看，index有点像spss里的casenum或者R里自动在表格最左列生成的序号

也可以自己指定index的label：

```python
obj2 = pd.Series([4, 7, -5, 3], index=['d', 'b', 'a', 'c'])
obj2
```

```python
obj2.index
```

#### 读取
可以用index的label来选择：

```python
obj2['a']
```

```python
obj2[['c', 'a', 'd']]
```

依照一定条件读取

```python
obj2[obj2 > 0]
```

#### 改变

```python
obj2['d'] = 6
obj2
```

```python
obj2.index = ['Bob', 'Steve', 'Jeff', 'Ryan']
obj2
```

#### 运算

```python
obj2 * 2
```

```python
import numpy as np
np.exp(obj2)
```

#### 找特定的索引是否存在

```python
'b' in obj2
```

#### 对齐并执行运算

## DataFrame
DataFrame就像SPSS的数据文件、R的数据框，SAS的dataset

### 用dict构建DataFrame

```python
data = {'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'], 
        'year': [2000, 2001, 2002, 2001, 2002, 2003], 
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data)

frame
```

```python
frame.head()
```

可以改变变量的顺序

```python
pd.DataFrame(data, columns=['year', 'state', 'pop'])
```

```python
frame2 = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'], 
                      index=['one', 'two', 'three', 'four', 'five', 'six'])
```

```python
frame2
```

```python
pop = {'Nevada': {2001: 2.4, 2002: 2.9},
       'Ohio': {2000: 1.5, 2001: 1.7, 2002: 3.6}}
frame3 = pd.DataFrame(pop)
frame3
```

```python
pdata = {'Ohio': frame3['Ohio'][:-1],
         'Nevada': frame3['Nevada'][:2]}
pd.DataFrame(pdata)
```

### 命名和单元格值

```python
frame3.index.name = 'year'; frame3.columns.name = 'state'
frame3
```

```python
frame3.values
```

```python
frame2.values
```

### 读取任意列

```python
frame2['state']
```

```python
frame2.year
```

### 删除任意列

```python
del frame2['year']
```

```python
frame2.columns
```

### 读取任意行：.loc[]

```python
frame2.loc['three']
```

### 改变DataFrame的值

```python
frame2['debt'] = 16.5
frame2
```

```python
frame2['debt'] = np.arange(6.)
frame2
```

### series合进DataFrame
把series赋给DataFrame，会按DataFrame的index来合并，不够的地方用缺失数据来表示：

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
frame2['debt'] = val
frame2
```

如果列不存在，赋值会创建一个新列

```python
frame2['eastern'] = frame2.state == 'Ohio'
frame2
```
